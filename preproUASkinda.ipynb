# FINDING K

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("DatasetUASFiltered.csv")
df_jokowi = df[df['Era'] == "Jokowi"]['stemmed'].to_list()
df_ahok = df[df['Era'] == "Ahok"]['stemmed'].to_list()
df_anies = df[df['Era'] == "Anies"]['stemmed'].to_list()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vec = TfidfVectorizer(max_df=0.95, min_df = 2, sublinear_tf=True)
X_jokowi = tfidf_vec.fit_transform(df_jokowi)
X_ahok = tfidf_vec.fit_transform(df_ahok)
X_anies = tfidf_vec.fit_transform(df_anies)
```

## Jokowi

```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

# Standarisasi data jika diperlukan
# scaled_data = preprocessing.StandardScaler().fit_transform(X_jokowi.toarray())

# Inisialisasi PCA dengan jumlah komponen yang sesuai
pca = PCA(n_components=min(X_jokowi.shape[0], X_jokowi.shape[1]))

# Fit PCA ke data TF-IDF
pca.fit(X_jokowi.toarray())

# Hitung varians yang dijelaskan dan kumulatif varians yang dijelaskan
explained_variance = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance)

# Visualisasi kumulatif varians yang dijelaskan
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance for PCA on TF-IDF')
plt.show()
```

```python
# Select the number of components based on the desired explained variance threshold
desired_explained_variance = 0.95
num_components = np.argmax(cumulative_explained_variance >= desired_explained_variance) + 1

# Output the selected number of components
print(f"Number of components to explain {desired_explained_variance:.2f} variance: {num_components}")

# Fit PCA with the selected number of components
pca = PCA(n_components=num_components)
PCA_result_jokowi = pca.fit_transform(X_jokowi.toarray())
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
cv_vec = CountVectorizer()
tfid_vec = TfidfVectorizer(max_df=0.96, min_df = 2, sublinear_tf=True)
```

```python
top_jokowi = []
for i in range(36):
    curr_top = []
    cluster_list = df_jokowi[df_jokowi["cluster"] == i]['stemmed'].to_list()
    X_curr = cv_vec.fit_transform(cluster_list)
    sum_words = X_curr.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in cv_vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    for term in range(15):
        curr_top.append(words_freq[term][0])
    top_jokowi.append(curr_top)
```

```python
top_jokowi_cluster = pd.DataFrame({'Cluster': [i for i in range(36)],
                                   'Top Terms': top_jokowi})
top_jokowi_cluster 
```

```python
top_ahok = []
for i in range(30):
    curr_top = []
    cluster_list = df_ahok[df_ahok["cluster"] == i]['stemmed'].to_list()
    X_curr = cv_vec.fit_transform(cluster_list)
    sum_words = X_curr.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in cv_vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    for term in range(15):
        curr_top.append(words_freq[term][0])
    top_ahok.append(curr_top)
```

```python
top_ahok_cluster = pd.DataFrame({'Cluster': [i for i in range(30)],
                                   'Top Terms': top_ahok})
top_ahok_cluster 
```

```python
top_anies = []
for i in range(30):
    curr_top = []
    cluster_list = df_anies[df_anies["cluster"] == i]['stemmed'].to_list()
    X_curr = cv_vec.fit_transform(cluster_list)
    sum_words = X_curr.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in cv_vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    for term in range(15):
        curr_top.append(words_freq[term][0])
    top_anies.append(curr_top)
```
